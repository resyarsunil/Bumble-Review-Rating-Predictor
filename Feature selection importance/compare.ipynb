```python
import pandas as pd
import numpy as np
import time

from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

from sklearn.pipeline import Pipeline

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer

from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_auc_score
)

import matplotlib.pyplot as plt

from sklearn.base import BaseEstimator, TransformerMixin

```

# Dataset1: Before FEATURE SELECTION

```python
df1 = pd.read_csv(
    r"C:\Users\surya\Downloads\data (1).csv"
)

x = df1.drop(['diagnosis', 'Unnamed: 32',"id"], axis=1)
y = df1['diagnosis']





# TRAIN TEST SPLIT

x_train, x_test, y_train, y_test = train_test_split(

    x,
    y,

    test_size=0.2,

    stratify=y,

    random_state=42
)


# MODELS

models = {

    "Random Forest": RandomForestClassifier(
        n_estimators=200,
        random_state=42
    ),

    "Naive Bayes": GaussianNB()
}


# CROSS VALIDATION


cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
```

```python
# BEFORE FEATURE SELECTION





# MODEL LOOP

for model_name, model in models.items():

    print(f"MODEL : {model_name}")
    print("BEFORE FEATURE SELECTION")


    # ========================================================
    # PIPELINE
    # ========================================================

    pipeline = Pipeline([

        ("imputer",
         SimpleImputer(strategy="median")),

        ("scaler",
         StandardScaler()),

        ("model", model)
    ])


    # ========================================================
    # CROSS VALIDATION
    # ========================================================

    cv_scores = cross_val_score(

        pipeline,

        x_train,
        y_train,

        cv=cv,

        scoring='accuracy',

        n_jobs=-1
    )


    # ========================================================
    # TRAINING
    # ========================================================

    start_time = time.time()

    pipeline.fit(x_train, y_train)

    end_time = time.time()

    training_time = end_time - start_time


    # ========================================================
    # PREDICTIONS
    # ========================================================

    y_pred = pipeline.predict(x_test)

    y_prob = pipeline.predict_proba(x_test)[:, 1]


    # ========================================================
    # METRICS
    # ========================================================

    accuracy = accuracy_score(y_test, y_pred)

    auc_score = roc_auc_score(y_test, y_prob)

    cm = confusion_matrix(y_test, y_pred)


    # ========================================================
    # PRINT RESULTS
    # ========================================================

    print(f"\nMean CV Accuracy     : {cv_scores.mean():.4f}")

    print(f"CV Standard Deviation: {cv_scores.std():.4f}")

    print(f"Test Accuracy        : {accuracy:.4f}")

    print(f"ROC-AUC Score        : {auc_score:.4f}")

    print(f"Training Time        : {training_time:.2f} sec")

    print("\nClassification Report:\n")

    print(classification_report(y_test, y_pred))


```

# Correlation feature selection

```python


feature_names = x.columns.tolist()





# ============================================================
# CORRELATION FEATURE SELECTION
# ============================================================

class CorrelationFeatureSelector(
    BaseEstimator,
    TransformerMixin
):

    def __init__(self, threshold=0.90):

        self.threshold = threshold


    def fit(self, X, y=None):

        # CONVERT TO DATAFRAME
        X = pd.DataFrame(
            X,
            columns=feature_names
        )

        corr_matrix = X.corr().abs()

        upper = corr_matrix.where(

            np.triu(
                np.ones(corr_matrix.shape),
                k=1
            ).astype(bool)
        )

        self.to_drop_ = [

            column for column in upper.columns

            if any(upper[column] > self.threshold)
        ]

        self.selected_features_ = [

            column for column in X.columns

            if column not in self.to_drop_
        ]

        return self


    def transform(self, X):

        # KEEP FEATURE NAMES
        X = pd.DataFrame(
            X,
            columns=feature_names
        )

        return X[self.selected_features_]





# ============================================================
# MODEL LOOP
# ============================================================

for model_name, model in models.items():


    print(f"MODEL : {model_name}")

    print("AFTER CORRELATION FEATURE SELECTION")



    # ========================================================
    # PIPELINE
    # ========================================================

    pipeline = Pipeline([

        ("imputer",
         SimpleImputer(strategy="median")),

        ("scaler",
         StandardScaler()),

        ("feature_selection",
         CorrelationFeatureSelector(threshold=0.90)),

        ("model", model)
    ])


    # ========================================================
    # CROSS VALIDATION
    # ========================================================

    cv_scores = cross_val_score(

        pipeline,

        x_train,
        y_train,

        cv=cv,

        scoring='accuracy',

        n_jobs=-1
    )


    # ========================================================
    # TRAINING
    # ========================================================

    start_time = time.time()

    pipeline.fit(x_train, y_train)

    end_time = time.time()

    training_time = end_time - start_time


    # ========================================================
    # SELECTED FEATURES
    # ========================================================

    selected_features = pipeline.named_steps[
        'feature_selection'
    ].selected_features_


    print(f"\nSelected Features ({len(selected_features)}):")

    print(selected_features)


    # ========================================================
    # PREDICTIONS
    # ========================================================

    y_pred = pipeline.predict(x_test)

    y_prob = pipeline.predict_proba(x_test)[:, 1]


    # ========================================================
    # METRICS
    # ========================================================

    accuracy = accuracy_score(y_test, y_pred)

    auc_score = roc_auc_score(y_test, y_prob)

    cm = confusion_matrix(y_test, y_pred)


    # ========================================================
    # PRINT RESULTS
    # ========================================================

    print(f"\nMean CV Accuracy     : {cv_scores.mean():.4f}")

    print(f"CV Standard Deviation: {cv_scores.std():.4f}")

    print(f"Test Accuracy        : {accuracy:.4f}")

    print(f"ROC-AUC Score        : {auc_score:.4f}")

    print(f"Training Time        : {training_time:.2f} sec")


    print("\nClassification Report:\n")

    print(classification_report(y_test, y_pred))





    # ========================================================
    # FEATURE IMPORTANCE
    # ========================================================

    plt.figure(figsize=(10, 6))


    # --------------------------------------------------------
    # RANDOM FOREST FEATURE IMPORTANCE
    # --------------------------------------------------------

    if model_name == "Random Forest":

        importances = pipeline.named_steps[
            'model'
        ].feature_importances_


        feature_importance_df = pd.DataFrame({

            'Feature': selected_features,
            'Importance': importances

        })


    # --------------------------------------------------------
    # NAIVE BAYES FEATURE IMPORTANCE
    # --------------------------------------------------------

    elif model_name == "Naive Bayes":

        nb_model = pipeline.named_steps['model']


        # MEAN DIFFERENCE IMPORTANCE
        importance = np.abs(

            nb_model.theta_[1]
            -
            nb_model.theta_[0]
        )


        feature_importance_df = pd.DataFrame({

            'Feature': selected_features,
            'Importance': importance

        })


    # ========================================================
    # SORT FEATURES
    # ========================================================

    feature_importance_df = feature_importance_df.sort_values(

        by='Importance',
        ascending=False
    )


    print("\nFeature Importance:\n")

    print(feature_importance_df)


    # ========================================================
    # FEATURE IMPORTANCE VISUALIZATION
    # ========================================================

    plt.figure(figsize=(10, 6))

    plt.barh(

        feature_importance_df['Feature'],

        feature_importance_df['Importance']
    )

    plt.xlabel("Feature Importance")

    plt.ylabel("Features")


    plt.title(

        f"{model_name} Feature Importance\n"
        f"After Correlation Feature Selection"
    )


    # MOST IMPORTANT FEATURE AT TOP
    plt.gca().invert_yaxis()

    plt.tight_layout()

    plt.show()
```

```python


from sklearn.feature_selection import RFE











# ============================================================
# NUMBER OF FEATURES TO SELECT
# ============================================================

n_features_to_select = 10


# ============================================================
# MODEL LOOP
# ============================================================

for model_name, model in models.items():

    print(f"MODEL : {model_name}")
    print("AFTER FEATURE SELECTION USING RFE")


    # ========================================================
    # RFE ESTIMATOR
    # ========================================================

    rfe_estimator = RandomForestClassifier(
        n_estimators=100,
        random_state=42
    )


    # ========================================================
    # PIPELINE
    # ========================================================

    pipeline = Pipeline([

        ("imputer",
         SimpleImputer(strategy="median")),

        ("scaler",
         StandardScaler()),

        # FEATURE SELECTION INSIDE PIPELINE
        # NO DATA LEAKAGE

        ("feature_selection",

         RFE(

             estimator=rfe_estimator,

             n_features_to_select=n_features_to_select
         )),

        ("model", model)
    ])


    # ========================================================
    # CROSS VALIDATION
    # ========================================================

    cv_scores = cross_val_score(

        pipeline,

        x_train,
        y_train,

        cv=cv,

        scoring='accuracy',

        n_jobs=-1
    )


    # ========================================================
    # TRAINING
    # ========================================================

    start_time = time.time()

    pipeline.fit(x_train, y_train)

    end_time = time.time()

    training_time = end_time - start_time


    # ========================================================
    # SELECTED FEATURES
    # ========================================================

    selected_mask = pipeline.named_steps[
        'feature_selection'
    ].support_

    selected_features = x.columns[selected_mask]


    # ========================================================
    # PREDICTIONS
    # ========================================================

    y_pred = pipeline.predict(x_test)

    y_prob = pipeline.predict_proba(x_test)[:, 1]


    # ========================================================
    # METRICS
    # ========================================================

    accuracy = accuracy_score(y_test, y_pred)

    auc_score = roc_auc_score(y_test, y_prob)

    cm = confusion_matrix(y_test, y_pred)


    # ========================================================
    # PRINT RESULTS
    # ========================================================

    print(f"\nSelected Features ({len(selected_features)}):")

    print(selected_features.tolist())


    print(f"\nMean CV Accuracy     : {cv_scores.mean():.4f}")

    print(f"CV Standard Deviation: {cv_scores.std():.4f}")

    print(f"Test Accuracy        : {accuracy:.4f}")

    print(f"ROC-AUC Score        : {auc_score:.4f}")

    print(f"Training Time        : {training_time:.2f} sec")


    print("\nClassification Report:\n")

    print(classification_report(y_test, y_pred))


   
```
