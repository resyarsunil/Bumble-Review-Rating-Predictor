```python
import numpy as np

# Initial population
population_2006 = 30

# Growth rate (10%)
growth_rate = 0.10

# Years from 2006 to 2026
years = list(range(2006, 2027))

# List to store population values
populations = []

# Calculate yearly population
population = population_2006

for year in years:
    populations.append(population)
    population = population * (1 + growth_rate)

# Mean population
mean_population = np.mean(populations)

# Standard deviation
std_population = np.std(populations)

print("Years:", years)
print("Population values:", populations)

print("\nMean Population:", mean_population)
print("Standard Deviation:", std_population)
```

```python
import math

# Population size
N = 1200000

# 95% confidence level -> Z value
Z = 1.96

# Population proportion (maximum variability)
p = 0.5

# Margin of error (5%)
e = 0.05

# Initial sample size
n = (Z**2 * p * (1 - p)) / (e**2)

# Finite Population Correction (FPC)
adjusted_n = n / (1 + ((n - 1) / N))

print("Initial Sample Size:", math.ceil(n))
print("Adjusted Sample Size:", math.ceil(adjusted_n))
```
