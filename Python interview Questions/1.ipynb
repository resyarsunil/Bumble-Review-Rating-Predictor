# 1. Write a Python program to count the number of consonants in a given string.

```python
def count_consonants(text):
    vowels = "aeiou"
    count = 0
    for char in text.lower():
        if char.isalpha() and char not in vowels:
            count +=1
    return count
```

```python
print(count_consonants("Hello world"))
```

# 2. Write a Python program to find the first non-repeating character in a string.

```python
def first_non_repeating_char(text):
    # STEP 1: CONVERT TEXT TO LOWERCASE
    lower_text = text.lower()

    # STEP 2: COUNT THE FREQUENCY OF EACH CHARACTER
    char_count = {}
    for char in lower_text:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    # STEP 3: FIND THE FIRST CHARACTER WITH COUNT 1
    for char in lower_text:
        if char_count[char] == 1:
            return char

    return None


```

```python
print(first_non_repeating_char("SWISS"))
```

# 3.Write a Python program to remove all duplicate characters from a string.

```python
def remove_duplicates(text):
    result = ""

    # STEP 1: IF CHARACTER NOT IN RESULT, ADD IT
    for char in text:
        if char not in result:
            result += char

    return result

```

```python
print(remove_duplicates("MALAYALAM"))
```

# 4. Write a Python program to count the frequency of each character in a string.

```python
def count_char_frequency(text):
    frequency = {}
    for char in text:

        # STEP 1: IF CHAR IN DICTIONARY, INCREASE COUNT
        if char in frequency:
            frequency[char] += 1

        # STEP 2: IF CHAR NOT IN DICTIONARY, INITIALIZE COUNT TO 1
        else:
            frequency[char] = 1

    return frequency
```

```python
 print(count_char_frequency("MALAYALAM"))
```
