# 1. Write a Python program to check whether a string contains only digits.

```python
def is_only_digits(text):

    #step 1: checks each character
    for char in text:
        if not char.isdigit():
            return False
    return f"contains only digits"
```

# 2. Write a Python program to find the longest word in a string.

```python
def find_longest_word(text):

    words = text.split()
    longest_word = ""

    for word in words:
        #if the curret word is longer than our previous longest, update it
        if len(word) > len(longest_word):
            longest_word = word
    return longest_word
```

```python
print(find_longest_word("The quick brown fox jumped over the lazy dog"))
```

# 3. Write a Python program to capitalize the first and last character of each word in a string.

```python
def capitalize_ends(text):

    words = text.split()
    result = []

    for word in words:
        if len(word) > 1:

            #capitalize first, keep middle, capitalize last

            new_word = word[0].upper() + word[1:-1] + word[-1].upper()
            result.append(new_word)
        else:
            result.append(word.upper())

    return " ".join(result)
```

```python
print(capitalize_ends("hello world! i am a coder."))
```

# 4. Draw a right-angled triangle pattern using stars.

```python
def right_traingle(n):

    for i in range(1, n+1):
        print("*"*i)
```

```python
print(right_traingle(5))
```

# 5. Draw an inverted right-angled triangle pattern.

```python
def inverted_triangle(n):

    for i in range(n, 0, -1):
        print("*" * i)
```

```python
print(inverted_triangle(5))
```

# 1 What is Machine Learning?

Machine Learning is a branch of Artificial Intelligence that allows computers to learn from data and improve their performance over time without being explicitly programmed. It identifies patterns and relationships within the data. Machine learning models can make predictions, classify information, or detect anomalies. It is widely used in applications such as recommendation systems, speech recognition, and medical diagnosis.

# 2 What are the main types of Machine Learning?

The main types of Machine Learning are supervised learning, unsupervised learning, and reinforcement learning. Supervised learning uses labeled data to train models to predict outcomes. Unsupervised learning finds patterns and structures in unlabeled data. Reinforcement learning trains agents to make decisions using a trial-and-error approach, rewarding correct actions and penalizing wrong actions.

# 3 What is the difference between Classification and Regression?

Classification predicts discrete categories, such as “spam” or “not spam,” while regression predicts continuous numerical values, like house prices or temperatures. Classification assigns inputs to predefined classes, whereas regression models relationships between variables to forecast numbers. Both are types of supervised learning used for different prediction tasks. Choosing the correct approach depends on whether the output is categorical or numerical.

# 4 What is a feature in Machine Learning?

A feature is an individual measurable property or characteristic of the data used for making predictions. Features are the inputs that a model uses to learn patterns and relationships. They can be numerical, categorical, or derived from raw data. Selecting relevant and high-quality features is critical for building an accurate machine learning model.

# 5 What is a label or target variable?

A label, also called a target variable, is the output a machine learning model aims to predict. It represents the answer or outcome corresponding to each input in the dataset. In supervised learning, the model learns the relationship between features and the label. Correctly defining the label is essential for effective model training and accurate predictions.
