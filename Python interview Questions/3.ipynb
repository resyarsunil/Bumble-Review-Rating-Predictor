# 1. Find the Factorial of a Number

```python
def find_factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i


    return result
```

```python
print(find_factorial(5))  
```

# 2. Find the Largest and Smallest Number in a List

```python
def find_min_max(numbers):
    if not numbers:
        return None
    
    smallest = numbers[0]
    largest = numbers[0]

    for num in numbers:
        if num<smallest:
            smallest = num
        if num>largest:
            largest = num

    return (smallest, largest)
```

```python
print(find_min_max([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]))
```

# 3. Fibonacci Sequence up to N terms

```python
def fibonacci_sequence(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b
```

```python
print(fibonacci_sequence(5))
```

# 4. Create a Square Frame Pattern

```python
def square_frame(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == 0 or j == n-1:
                print("*", end="")
            else:
                print(" ", end="")
        print()
```

```python
print(square_frame(5))
```

# 5. Sum of Digits in a Number

```python
def sum_of_digits(n):
    total = 0
    for digit in str(n):
        total +=int(digit)
    return total
```

```python
print(sum_of_digits(12345)) 
```

# 1. What are Overfitting and Underfitting?

Overfitting occurs when a machine learning model learns the training data too well, capturing noise and performing poorly on new data. Underfitting happens when a model is too simple to capture the patterns in the data, resulting in poor performance on both training and testing sets. Overfitting leads to high variance, while underfitting leads to high bias. Balancing model complexity and using techniques like cross-validation helps prevent both problems.

# 2. What is the difference between Supervised and Unsupervised Learning?

Supervised learning uses labeled data to train models to predict outcomes or classify data. Unsupervised learning works with unlabeled data and finds hidden patterns or groupings. Supervised learning outputs a specific target variable, while unsupervised learning does not. The choice depends on whether labeled outcomes are available in the dataset.

# 3. What is a training dataset and a testing dataset? Why is data splitting important?

A training dataset is used to teach the model and adjust its parameters. A testing dataset is used to evaluate the model’s performance on unseen data. Splitting data ensures that the model generalizes well and does not memorize the training data. This practice helps detect overfitting and provides a realistic measure of accuracy.

# 4. What is feature scaling and why is it needed in some algorithms?

Feature scaling is the process of normalizing or standardizing input features so they have similar ranges. It is important for algorithms that use distance calculations, such as K-Nearest Neighbors or gradient-based methods. Without scaling, features with larger values can dominate the learning process. Scaling ensures faster convergence and more accurate models.

# 5. How does a Linear Regression model work?

Linear Regression models the relationship between input features and a continuous target variable using a straight line. It calculates coefficients for each feature to minimize the difference between predicted and actual values. The model predicts outputs by applying these coefficients to the input features. Linear Regression is simple, interpretable, and widely used for numerical predictions.

The assumptions of Linear Regression are:

There is a linear relationship between the independent variables (features) and the dependent variable (target).

The residuals (errors) are normally distributed and have a mean of zero.

The variance of the residuals is constant across all levels of the independent variables, known as homoscedasticity.

The observations are independent, and there is no multicollinearity (independent variables are not highly correlated with each other).
